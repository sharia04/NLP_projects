```python
import pandas as pd 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# sample dataset
data = pd.DataFrame({
    'Movie': ['Movie A','Movie B','Movie c'],
    'Description':[
        'Action and adaventure in space',
        'Romantic comedy with friends',
        'Sci-fi thriller in space'
    ]
})
```

```python
#TF-IDE vectorization
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(data['Description'])
```

```python
#Cosine similarity
cos_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
#Recommend movie similar to Movie A
movie_idx = 0
similar_movies = list(enumerate(cos_sim[movie_idx]))
similar_movies = sorted(similar_movies, key=lambda x: x[1],reverse= True)
print("Recommendation for Movie A:")
for idx, score in similar_movies[1:]:
    print(data['Movie'][idx], "Score:", score)
```
