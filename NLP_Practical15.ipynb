```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
import heapq
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
text = """
Artificial Intelligence (AI) is a branch of computer science that aims to create intelligent machine.
It has application in various fields such as healthcare, finance, education, and autonomous vehicles.
Machine Learning , a sunset of AI, allows system to learn from data without explicit programing.
Natural Language proccesing enables to machine to understand and interpret human language.
AI continues to grow rapidly and impact everyday life in multiple ways.
"""
```

```python
stop_words = set(stopwords. words("english"))
word = word_tokenize(text.lower())
word_frequencies = {}

for word in word:
    if word not in stop_words and word.isalpha():
        word_frequencies[word] = word_frequencies.get(word, 0)+1

sentences = sent_tokenize(text)
sentence_score ={}
for sent in sentences:
    for word in word_tokenize(sent.lower()):
        if word in word_frequencies:
            sentence_score[sent] = sentence_score.get(sent, 0) +word_frequencies[word]

summary_sentences = heapq.nlargest(3, sentence_score, key=sentence_score.get)
summary = " ".join(summary_sentences)
print("Summary:\n", summary)
```
