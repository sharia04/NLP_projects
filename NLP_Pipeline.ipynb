```python
text="Artificial Tntiligence is Changing the world Google is located in USA"
```

```python
#segmentation
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(text)
print("segmentation:",sentences)
```

```python
#tokenization
from nltk.tokenize import word_tokenize
tokens=word_tokenize(text)
print("tokenization",tokens)
```

```python
#stemming
from nltk.stem import PorterStemmer
ps=PorterStemmer()
stemmed= [ps.stem(w) for w in tokens]
print("Stemming", stemmed)
```

```python
#lemmatization 
from nltk.stem import WordNetLemmatizer
lemm= WordNetLemmatizer()
lemmatized=[lemm.lemmatize(w) for w in tokens]
print("Lemmatization", lemmatized)
```

```python
#POS_Tagging
pos_tags = nltk.pos_tag(tokens)
print("POS Tagging:",pos_tags)
```

```python
#NER
named_entities = nltk.ne_chunk(pos_tags)
print("Named.Entities:",named_entities)
```
