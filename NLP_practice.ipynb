```python
from nltk import pos_tag
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
sentence = "The dogs are running"
tokens = word_tokenize(sentence)
tagged = pos_tag(tokens)

lemmatized_word = {lemmatizer.lemmatize(
    word, pos ='v' if tag.startswith('v') else 'n') for word, tag in tagged}
print(lemmatized_word)
#Task : Apply for sentence: "I am Studing book"
```

```python
import nltk 
from nltk.tokenize import word_tokenize
from nltk import pos_tag
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
text = "NLTK is a powerful library for natural language processing."
words = word_tokenize(text)
```

```python
pos_tags = pos_tag(words)
```

```python
print("Original Taxt:")
print(text)

print("\nPOS Tagging Result:")
for word, pos_tags in pos_tags:
    print(f"{word}: {pos_tags}")
# the quick brown for jumps over the lazy dog
```
